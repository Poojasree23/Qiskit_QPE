```python
!pip install qiskit qiskit-aer --quiet

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit_aer import AerSimulator, noise
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------------------------
# 1. UNITARY: U |1> = e^(2πiθ)|1>
# ---------------------------------------------

def unitary(theta):
    """Returns a U gate with phase e^(2πiθ)."""
    return np.array([[1, 0],
                     [0, np.exp(2j*np.pi*theta)]])

# ---------------------------------------------
# QPE Circuit Builder
# ---------------------------------------------
def qpe(theta, counting_qubits=3, noisy=False):
    total_qubits = counting_qubits + 1
    qc = QuantumCircuit(total_qubits)

    # Counting qubits → Hadamard
    qc.h(range(counting_qubits))

    # Target qubit = |1>
    qc.x(counting_qubits)

    # Apply controlled-U^2^k
    U = unitary(theta)
    for k in range(counting_qubits):
        power = 2**k
        U_power = np.linalg.matrix_power(U, power)
        qc.unitary(U_power, [k, counting_qubits], label=f"U^{power}")

    # Inverse QFT
    for qubit in range(counting_qubits//2):
        qc.swap(qubit, counting_qubits-1-qubit)

    for j in range(counting_qubits):
        for m in range(j):
            qc.cp(-np.pi/float(2**(j-m)), m, j)
        qc.h(j)

    qc.measure_all()

    sim = AerSimulator()

    if noisy:
        # Noise model with depolarizing noise
        error = noise.depolarizing_error(0.05, 1)
        noise_model = noise.NoiseModel()
        noise_model.add_all_qubit_quantum_error(error, ['unitary'])

        result = sim.run(qc, shots=2048, noise_model=noise_model).result()

    else:
        result = sim.run(qc, shots=2048).result()

    return qc, result.get_counts()


# ---------------------------------------------
# Helper to get binary and decimal results
# ---------------------------------------------
def analyze(theta, counts, n_qubits):
    measured = max(counts, key=counts.get)
    measured_decimal = int(measured, 2) / (2**n_qubits)

    print(f"\nTheta = {theta}")
    print(f"Most frequent output: {measured}")
    print(f"Measured decimal phase = {measured_decimal}")
    print(f"Expected phase = {theta}\n")


# ---------------------------------------------
# RUN FOR MULTIPLE THETA VALUES
# ---------------------------------------------

theta_values = [0.25, 0.375, 0.5]
counting_qubits = 4

for theta in theta_values:
    qc, counts = qpe(theta, counting_qubits)
    analyze(theta, counts, counting_qubits)

    plt.figure()
    plt.bar(counts.keys(), counts.values())
    plt.title(f"QPE Output Histogram (θ = {theta})")
    plt.show()

    print("\nInverse QFT structure:")
    display(qc.draw('text'))

# ---------------------------------------------
# NOISE SIMULATION EXAMPLE
# ---------------------------------------------
theta = 0.375
qc_noisy, noisy_counts = qpe(theta, counting_qubits, noisy=True)

plt.figure()
plt.bar(noisy_counts.keys(), noisy_counts.values())
plt.title(f"Noisy QPE Output (θ = {theta})")
plt.show()

print("\nNOISE SIMULATION RESULT:")
analyze(theta, noisy_counts, counting_qubits)
```
